<h3>1. Categorical Data

<h5> a. Countplot

```python
ti_ds['Survived'].unique()
```

```python
ti_ds['Survived'].value_counts()
```

```python
sns.countplot(data = ti_ds, x = 'Survived')
```

```python
sns.countplot(data = ti_ds, x = 'Pclass')
```

```python
sns.countplot(data = ti_ds, x = 'Sex')
```

```python
sns.countplot(data = ti_ds, x = 'Embarked')
```

<h5>b. PieChart

Pie chart is used to find the value in percentage 

```python
ti_ds['Survived'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
ti_ds['Pclass'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
ti_ds['Sex'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```
